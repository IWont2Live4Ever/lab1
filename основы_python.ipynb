#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def func(input):
    vowels = "aeiou"
    return sum(1 for char in input.lower() if char in vowels)

assert func("") == 0              # пустая строка
assert func("aeiou") == 5         # a, e, i, o, u
assert func("bcdfg") == 0         # -
assert func("func") == 1          # u
assert func("abracadabra") == 5   # 5a
assert func("   cen") == 1        # e
assert func("AbCdEf") == 2        # A, E

print("тесты пройдены")
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def func(input):
    return len(set(input)) == len(input)

assert func("") == True          # пустая строка
assert func("a") == True         # один символ
assert func("abc") == True       # все символы уникальны
assert func("aa") == False       # повторяющиеся символы
assert func("abca") == False     # повтор в конце
assert func("a1b2c3") == True    # уникальные буквы и цифры
assert func("a a") == False      # пробел и повторяющаяся буква

print("тесты пройдены")
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def func(input):
    return bin(input)[2:].count('1')

assert func(0) == 0    # 0
assert func(1) == 1    # 1
assert func(2) == 1    # 10
assert func(3) == 2    # 11
assert func(5) == 2    # 101
assert func(7) == 3    # 111
assert func(20) == 2   # 10100

print("тесты пройдены")
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    steps = 0
    while n >= 10:
        product = 1
        for digit in str(n):
            product *= int(digit)
        n = product
        steps += 1
    return steps

assert magic(39) == 3   # 3*9=27 → 2*7=14 → 1*4=4
assert magic(4) == 0    # уже одна цифра
assert magic(999) == 4  # 9*9*9=729 → 7*2*9=126 → 1*2*6=12 → 1*2=2
assert magic(10) == 1   # 1*0=0
assert magic(25) == 2   # 2*5=10 → 1*0=0
assert magic(123) == 1  # 1*2*3=6

print("тесты пройдены")
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if not pred:  # если пустые списки
        return 0.0
    n = len(pred)
    squaredErrors = [(p - t) ** 2 for p, t in zip(pred, true)]
    return sum(squaredErrors) / n

assert mse([], []) == 0.0                        # пустые списки
assert mse([1, 2, 3], [1, 2, 3]) == 0.0          # идентичные векторы
assert mse([1, 2, 3], [4, 5, 6]) == 9.0          # все разности по 3
assert mse([0, 0], [1, 1]) == 1.0                # разности по 1
assert mse([5], [5]) == 0.0                      # один элемент
assert mse([1, 2, 3, 4], [1, 2, 3, 5]) == 0.25   # последний отличается на 1

print("тесты пройдены")
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def func(n):
    if n <= 1:
        return ""
    
    factors = {}
    d = 2
    while d * d <= n:
        cnt = 0
        while n % d == 0:
            cnt += 1
            n //= d
        if cnt:
            factors[d] = cnt
        d += 1 if d == 2 else 2
    if n > 1:
        factors[n] = 1
    
    result_parts = []
    for prime in sorted(factors):
        power = factors[prime]
        if power == 1:
            result_parts.append(f"({prime})")
        else:
            result_parts.append(f"({prime}**{power})")
    return "".join(result_parts)

assert func(24) == "(2**3)(3)"
assert func(36) == "(2**2)(3**2)"
assert func(17) == "(17)"            # простое число
assert func(64) == "(2**6)"          # степень двойки
assert func(1) == ""                 # граничный случай

print("тесты пройдены")
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    if number <= 0:
        return "It is impossible"
    
    total = 0
    k = 0
    while total < number:
        k += 1
        total += k * k
        if total == number:
            return k
    return "It is impossible"

assert pyramid(1) == 1                     # 1 = 1**2
assert pyramid(5) == 2                     # 5 = 1**2 + 2**2
assert pyramid(14) == 3                    # 14 = 1**2 + 2**2 + 3**2
assert pyramid(30) == 4                    # 30 = 1**2 + 2**2 + 3**2 + 4**2
assert pyramid(55) == 5                    # 55 = 1**2 + 2**2 + 3**2 + 4**2 + 5**2
assert pyramid(2) == "It is impossible"    # нельзя представить

print("тесты пройдены")
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def func(number):
    s = str(number)
    n = len(s)
    if n % 2 == 1:
        mid = n // 2
        left_sum = sum(int(d) for d in s[:mid])
        right_sum = sum(int(d) for d in s[mid+1:])
    else:
        left_sum = sum(int(d) for d in s[:n//2 - 1])
        right_sum = sum(int(d) for d in s[n//2 + 1:])
    return left_sum == right_sum

assert func(5) == True          # одна цифра, средняя 5, слева и справа пусто → 0==0
assert func(12) == True         # две цифры, обе средние, пустые суммы → 0==0
assert func(121) == True        # нечётное, средняя 2, 1==1
assert func(1221) == True       # чётное, средние 2 и 2, 1==1
assert func(123) == False       # нечётное, средняя 2, 1 != 3
assert func(1234) == False      # чётное, средние 2 и 3, 1 != 4

print("тесты пройдены")
```
